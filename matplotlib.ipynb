```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

import numpy as np
```

## Diferentes plots
* Diagramas de dispersión o puntos.
* Diagramas de barras.
* Histograma.
* Diagramas de caja y bigotes.

#### Diagrama de Dispersión

```python
# Generar datos aleatorios para x e y
x = np.random.randn(100)
y = np.random.randn(100)

# Crear un diagrama de dispersión
fig, ax = plt.subplots()
ax.scatter(x, y)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Diagrama de Dispersión')

# Mostrar el gráfico
plt.show()
```

#### Diagrama de Barras

```python
# Generar datos aleatorios para el eje x e y
x = np.array(['A', 'B', 'C', 'D', 'E'])
y = np.random.rand(5)

# Crear un diagrama de barras
fig, ax = plt.subplots()
ax.bar(x, y)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Diagrama de Barras')

# Mostrar el gráfico
plt.show()
```

#### Histograma

```python
# Generar datos aleatorios para el eje x
x = np.random.randn(1000)

# Crear un histograma
fig, ax = plt.subplots()
ax.hist(x, bins=20)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma')

# Mostrar el gráfico
plt.show()
```

#### Diagrama de Caja

```python
# Generar datos aleatorios para el eje x
x = np.random.randn(1000)

# Crear un diagrama de caja
fig, ax = plt.subplots()
ax.boxplot(x)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_title('Diagrama de Caja y Bigotes')

# Mostrar el gráfico
plt.show()
```

```python
# Create the figure and axes objects
fig, ax = plt.subplots(figsize=(8,8))
# Set up the map projection
m = Basemap(llcrnrlon=-119, llcrnrlat=22, urcrnrlon=-64, urcrnrlat=49, projection='lcc', lat_1=33, lat_2=45, lon_0=-95, resolution='l', area_thresh=1000, ax=ax)

# Draw the coastlines and states
m.drawcoastlines(linewidth=0.5)
m.drawstates(linewidth=0.5)

# Fill in the land areas with green shading
m.etopo()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    {'metros': [1, 2, 3, 4, 5, 6, 7, 8],
     'ninos': [2, 6, 10, 5, 10, 3, 2, 2]}
)

weighted_median = np.average(df['metros'], weights=df['ninos'])
print(weighted_median)
```

```python
weighted_avg = (df['x'] * df['w']).sum() / df['w'].sum()

print(weighted_avg)
```

```python
std_w = np.sqrt(
    np.average(
        (df['metros'] - weighted_avg)**2,
        weights=df['ninos']
    )
)
print('Weighted Standard Deviation:', std_w)
```
